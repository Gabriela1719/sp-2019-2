```c++
#define BLOK 60
typedef struct {
    int pbr;
    char mjesto[50+1] = "";
    char opcina[50+1] = "";
}zapis;
```

```c++
void print(zapis tmp){
    cout 
        << tmp.pbr << "-"
        << tmp.mjesto << "-"
        << tmp.opcina << "\n";
}
```

```c++
int F;
```

**Prebaci podatke u neformatiranu datoteku  "mjesta.dat"**

```c++
{
    FILE *f, *fb;
    zapis pom;
    int br;
    f = fopen("mjesta.txt", "r");
    fb = fopen("mjesta.dat", "wb+");
    // PoÄetak datoteke
    fseek(f,0,SEEK_SET);
    fseek(fb,0,SEEK_SET);
    // Prebaci podatke
    br = 0;
    while(fscanf(f,"%[^,],%d,%[^\n] ",pom.mjesto, &pom.pbr, pom.opcina) != EOF){
        fwrite(&pom,sizeof(zapis),1,fb);
        br++;
        //print(pom);
    }
    F = br;
    fclose(f);
    fclose(fb);
    cout << "Ukupno: " << br << " zapisa";
}
```

##  Parametri

```c++
cout <<"Zapisa: "<< F;
cout <<"\nBlokova: " << ceil((float)F/BLOK);
cout <<"\nBlok: " << BLOK;
```

**4. PronaÄ‘i zapis**


```c++
const char *naziv = "Malinska";
//naziv = "Malinska";
```

```c++
int trazi(const char *naziv);
{
    zapis pom;
    FILE *f = fopen("mjesta.dat", "r");
    int br = 0;
    bool pronadjen = false;

    // PronaÄ‘i blok
    while(fread(&pom, sizeof(zapis),1, f)){
        //print(pom);
        if(strcmp(pom.mjesto, naziv)== 0)
            return 1;
        else if(strcmp(pom.mjesto, naziv)>0){
            pronadjen = true;
            break;
        }else{
            br++;
            fseek(f, br * BLOK * sizeof(zapis), SEEK_SET);
        }
    }
    // PronaÄ‘i zapis u bloku
    if(pronadjen){
        
        br -= 1;
        
        // Vrati se jedan blok unatrag
        fseek(f, br*BLOK*sizeof(zapis),SEEK_SET);
        
        // Ovdje napisati kod
        
        return 0;
    }else
        return 0;
}
```

#### 1. Napisati rekurzivnu funkciju koja rauna n-ti lan aritmetikog niza?

```c++
int aniz(int a0, int d, int n) {
   if (n == 0) return a0;
   else return d + aniz(a0, d, n-1);
}
```

#### 2. Napisati rekurzivnu funkciju koja vraa sumu znamenki zadanog broja?

```c++
int sum_of_digit(int n)
{
    if(n==0)
        return 0;
    int z = n%10;
    return z + sum_of_digit(n/10);
} 
  
```

```c++
{
    int broj = 345;
    print(sum_of_digit(broj));
}
```

```c++
int suma(int a, int b){
    if (a>b)
        return 0;
    return a * suma(a+1, b);
}
```

```c++
suma(2,6)
```

#### 3. Napisati rekurzivnu funkciju koja ispisiva niz u obrnutom redoslijedu?

```c++
void obrnuto(char niz[], int d) {
    if(d>0){
        print(niz[d-1]);
        obrnuto(niz,d-1);
    }
}
```

```c++
{
    char tmp[] = "ovo je recenica";
    int d = strlen(tmp);
    obrnuto(tmp, d);
}
```

```c++
void obrnuto2(char *niz)
{
    if(*niz == '\0')
        return;
    obrnuto2(niz+1);
    print(*niz);
}
```

```c++
{
    char tmp[] = "ovo je recenica";
    obrnuto2(tmp);
}
```

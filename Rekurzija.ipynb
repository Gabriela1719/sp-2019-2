#### Što će ispisati funkcija?

```c++
void pisi1 (int broj, int n) {
  broj++;
  if (broj > n) return;
  pisi1 (broj, n);
  print(broj);
}
```

#### Što će ispisati  funkcija?

```c++
void pisi3 (int *broj, int n) {
  (*broj)++;
  if (*broj > n) return;
  pisi3(broj, n);
  print(*broj);
}
```

```c++
{
    int broj = 0;
    pisi3(&broj, 5);
}
```

#### Što će ispisati  funkcija?

```c++
void pisi5 (int *broj, int n) {
  (*broj)--;
  if (*broj < 0) return;
  pisi5 (broj, n);
  print(*broj);
}
```
